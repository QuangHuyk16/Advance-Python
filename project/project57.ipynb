# Turtle Graphics with Python

Turtle Graphics là một công cụ thú vị để khám phá và tạo ra các hình ảnh động trong Python.
Mô-đun rùa cung cấp một môi trường trong đó rùa di chuyển xung quanh trên lưới 2 chiều. Rùa có vị trí, hướng (hướng rùa đang đối mặt) và nhiều trạng thái có thể xảy ra (rùa có thể vẽ các đường màu cụ thể khi chúng di chuyển hoặc không để lại dấu vết) và hành động (rẽ trái hoặc phải) di chuyển về phía trước hoặc phía sau.

Một đối tượng Rùa có nhiều phương pháp có thể được nhóm lại thành các gia đình.

```python
#Spiral Art (Rùa sẽ di chuyển theo hình xoắn ốc và thay đổi màu sắc của bút theo chu kỳ)

import turtle
colors=['red','purple','blue','green','orange','yellow']
t=turtle.Pen()
turtle.bgcolor('black')
for x in range(360):
    t.pencolor(colors[x%6])
    t.width(x/100 + 1)
    t.forward(x)
    t.left(59)
    t.speed(10000)
```

```python
#Mẫu cây fractal

import turtle as tu

roo = tu.Turtle()  # Đối tượng Turtle
wn = tu.Screen()  # Đối tượng màn hình
wn.bgcolor("black")  # Màu nền của màn hình
wn.title("Mẫu cây Fractal")
roo.left(90)  # Quay rùa 90 độ về bên trái
roo.speed(20)  # Đặt tốc độ của rùa


def draw(l):  # Hàm đệ quy với độ dài 'l' là tham số
    if l < 10:
        return
    else:
        roo.pensize(2)  # Đặt kích thước bút
        roo.pencolor("yellow")  # Đặt màu bút là màu vàng
        roo.forward(l)  # Di chuyển rùa về phía trước một khoảng 'l'
        roo.left(30)  # Quay rùa 30 độ về bên trái
        draw(3 * l / 4)  # Vẽ một hình fractal bên trái của đối tượng rùa 'roo' với 3/4 chiều dài của nó
        roo.right(60)  # Quay rùa 60 độ về bên phải
        draw(3 * l / 4)  # Vẽ một hình fractal bên phải của đối tượng rùa 'roo' với 3/4 chiều dài của nó
        roo.left(30)  # Quay rùa 30 độ về bên trái
        roo.pensize(2)
        roo.backward(l)  # Di chuyển rùa trở lại vị trí ban đầu
draw(20)  # Vẽ 20 lần

roo.right(90)
roo.speed(2000)


# Đệ quy
def draw(l):
    if l < 10:
        return
    else:
        roo.pensize(2)
        roo.pencolor("magenta")  # Màu magenta
        roo.forward(l)
        roo.left(30)
        draw(3 * l / 4)
        roo.right(60)
        draw(3 * l / 4)
        roo.left(30)
        roo.pensize(2)
        roo.backward(l)
draw(20)

roo.left(270)
roo.speed(2000)


# Đệ quy
def draw(l):
    if l < 10:
        return
    else:
        roo.pensize(2)
        roo.pencolor("red")  # Màu đỏ
        roo.forward(l)
        roo.left(30)
        draw(3 * l / 4)
        roo.right(60)
        draw(3 * l / 4)
        roo.left(30)
        roo.pensize(2)
        roo.backward(l)
draw(20)

roo.right(90)
roo.speed(2000)


# Đệ quy
def draw(l):
    if l < 10:
        return
    else:
        roo.pensize(2)
        roo.pencolor('#FFF8DC')  # Màu trắng
        roo.forward(l)
        roo.left(30)
        draw(6 * l / 7)
        roo.right(60)
        draw(6 * l / 7)
        roo.left(30)
        roo.pensize(2)
        roo.backward(l)
draw(60)

wn.exitonclick()
```

Kết hợp 

```python
import turtle

# Tạo một đối tượng turtle
t = turtle.Turtle()

# Đặt tốc độ cho turtle
t.speed(1)

# Vẽ một hình vuông
for _ in range(4):
    t.forward(100)  # Di chuyển turtle về phía trước 100 đơn vị
    t.right(90)     # Quay turtle 90 độ sang phải

# Kết thúc vẽ
turtle.done()
```

```python
import turtle as tu

# Khởi tạo đối tượng Turtle
roo = tu.Turtle()

# Đặt tốc độ vẽ
roo.speed(1)

# Vẽ tam giác
for _ in range(3):
    roo.forward(100)  # Di chuyển về phía trước 100 đơn vị
    roo.left(120)  # Quay trái 120 độ

# Đóng cửa sổ Turtle Graphics khi hoàn thành
tu.done()
```

```python
# Biểu đồ sunburst thể hiện cấu trúc dữ liệu phân cấp bằng cách sử dụng các vòng đồng tâm.
# Mỗi vòng biểu thị một cấp độ khác nhau của hệ thống phân cấp.
# Các phân đoạn trong mỗi vòng biểu thị các danh mục hoặc danh mục phụ.

# mình sẽ tạo một biểu đồ sunburst từ dữ liệu mẫu về hóa đơn và tiền tip tại các nhà hàng.

# ta sẽ gọi thư viện bằng cách import plotly.express và gán tên là px
# Thư viện này giúp tạo các biểu đồ trực quan một cách dễ dàng.
import plotly.express as px

#  Tập dữ liệu mẫu này chứa thông tin về hóa đơn và tiền tại các nhà hàng.
data_1 = px.data.tips()

# tạo biểu đồ sunburst và đặt tên biến là figure
figure = px.sunburst(data_1, path=["day", "sex",'time','size'], values="total_bill")
# - Các tham số:
#`data`: DataFrame chứa dữ liệu.
#`path`: list các cột trong Data để xác định cấu trúc của biểu đồ sunburst.
#`values`: Cột trong Data chứa giá trị để hiển thị trên biểu đồ. 
#   Ở đây, chúng ta sử dụng cột "total_bill".


# `figure.show()` : giúp hiển thị biểu đồ sunburst ra màn hình.
figure.show()
```

Ưu điểm:
- Trực quan hóa hệ thống phân cấp: Biểu đồ sunburst rất hữu ích khi bạn muốn minh họa cấu trúc dữ liệu phân cấp, đặc biệt là trong các mối quan hệ cha-con.
- Tỷ lệ miêu tả: Biểu đồ sunburst thể hiện tỷ lệ của các phân khúc trong tập dữ liệu, giúp hiểu về ý nghĩa tương đối.

Nhược điểm:
- Không phù hợp với dữ liệu lớn: Khi có quá nhiều phân đoạn, biểu đồ sunburst có thể trở nên khó đọc và không hiệu quả.
- Yêu cầu dữ liệu phân cấp theo nhóm cụ thể: Dữ liệu phải được sắp xếp theo kiểu phân cấp và mỗi cấp độ phân cấp phải có cột chuyên dụng.

Ưu và Nhược điểm của biểu đồ scatter_3d

Ưu điểm:

- Khám phá không gian 3D: Biểu đồ phân tán 3D cho phép bạn khám phá không gian ba chiều, giúp hiểu rõ hơn về sự phân bố của dữ liệu.

Nhược điểm:
- Phức tạp hóa trực quan: Biểu đồ phân tán 3D có thể trở nên phức tạp hơn so với biểu đồ phân tán 2D, đặc biệt khi có nhiều điểm dữ liệu. Điều này làm cho việc hiểu và trình bày dữ liệu trở nên khó khăn hơn.
- Khả năng hiển thị bị hạn chế: Không phải tất cả các trình duyệt hỗ trợ WebGL, do đó, việc hiển thị biểu đồ phân tán 3D có thể gặp khó khăn trên một số trình duyệt.


```python
figure_choropleth = px.choropleth(data,
                    # locations='iso_alpha': Xác định mã quốc gia để liên kết với các vùng trên bản đồ.
                    locations='iso_alpha',
                    # color='lifeExp': Màu sắc của các vùng sẽ phụ thuộc vào cột lifeExp
                    color='lifeExp',
                    # projection='natural earth': Sử dụng hình chiếu bản đồ thế giới theo tự nhiên.
                    projection='natural earth',
                    # height=700: Đặt chiều cao của biểu đồ là 700 pixel.
                    height=700,
                    # hover_data=data.columns: Hiển thị thông tin các cột khi di chuột qua các vùng trên bản đồ.
                    hover_data=data.columns,
                    # hover_name='country': Hiển thị tên quốc gia khi di chuột qua các vùng.
                    hover_name='country')
# Hiển thị bản đồ
figure_choropleth.show()
```

các nước tại nam phi có tuổi thọ trung bình rất thấp do:

Ưu điểm của biểu đồ choropleth:
- Hiệu quả trực quan: Biểu đồ choropleth giúp hiển thị một lượng lớn thông tin và các mẫu tổng quát về dữ liệu.
- Phù hợp cho nhóm khu vực: Biểu đồ này thường sử dụng các nhóm khu vực như quốc gia, tiểu bang hoặc lục địa, phù hợp cho việc so sánh dữ liệu giữa các khu vực.

Nhược điểm của biểu đồ choropleth:
- Lựa chọn khoảng giá trị: Cần chọn cẩn thận khoảng giá trị (interval) hoặc số lớp (class sizes) để tạo màu sắc cho các khu vực.
- Phụ thuộc vào kích thước khu vực: Mức độ tô màu phụ thuộc vào kích thước của các khu vực hành chính được chọn.

```python
map_fig = px.scatter_geo(data,  # tạo một biểu đồ phân tán dựa trên dữ liệu data 
                        # locations='iso_alpha' xác định cột trong data chứa mã ISO alpha-3
                        # của các quốc gia (ví dụ: “VNM” cho Việt Nam).
                        locations='iso_alpha',
                        animation_frame='year',animation_group='country',
                        # projection='orthographic' thiết lập kiểu dự án bản đồ là “orthographic” 
                        # (chế độ chiếu hình cầu).
                        projection='orthographic',
                        # color='continent' xác định cột trong data để tô màu các điểm dựa trên lục địa.
                        color='continent',
                        # opacity=0.8 đặt độ mờ của các điểm trên bản đồ.
                        opacity= 1,
                        # hover_name='country' hiển thị tên của quốc gia khi di chuột qua các điểm.
                        hover_name='country',
                        # hover_data=['gdpPercap', 'pop', 'year'] hiển thị thông tin bổ sung 
                        # khi di chuột qua các điểm.
                        hover_data=['gdpPercap','pop','year'])
map_fig.show()
```

Ưu điểm:
- Dễ hiểu và so sánh: Biểu đồ scatter_geo cho phép bạn thấy rõ xu hướng tăng giảm của dữ liệu trên bản đồ. Điều này giúp dễ dàng so sánh các giá trị địa lý.
- Tự động cập nhật: Khi dữ liệu thay đổi, biểu đồ scatter_geo sẽ tự động cập nhật vị trí của các điểm trên bản đồ.
- Phong phú về dạng biểu đồ: Bạn có thể tùy chỉnh biểu đồ scatter_geo để hiển thị các thông tin khác nhau, ví dụ như màu sắc, kích thước, văn bản, và biểu tượng.

Nhược điểm:
- Dễ bị nhầm lẫn: Vì biểu đồ scatter_geo sử dụng các biểu tượng (điểm) trên bản đồ, nên nó có thể dễ bị nhầm lẫn nếu có quá nhiều điểm trùng hợp.
- Không phù hợp với dữ liệu lớn: Khi có quá nhiều điểm, biểu đồ scatter_geo có thể trở nên khó đọc và không hiệu quả.

```python

data  = px.data.gapminder()
figure_scatter = px.scatter(data ,
            #  trục x là gdpPercap (GDP per capita), và trục y là lifeExp (tuổi thọ).
            x='gdpPercap',y='lifeExp',
            # animation_frame='year': Tạo hiệu ứng chuyển động theo năm,
            # với mỗi khung hình biểu diễn dữ liệu cho một năm khác nhau.
            # animation_group='country': Nhóm dữ liệu theo tên quốc gia để tạo hiệu ứng chuyển động cho từng quốc gia.
            animation_frame='year',animation_group='country',
            # vì trục y đag chạy vượt quá phía trên vì vậy mình sẽ tăng giới hạn trục y lên trong khoảng từ 20->100
            range_y=[20,100],
            # size='pop': Kích thước điểm trên biểu đồ phụ thuộc vào cột pop (dân số).
            size='pop',
            # color='continent': Màu sắc của điểm biểu diễn theo cột continent (lục địa).
            color='continent',
            # height=700: Đặt chiều cao của biểu đồ là 700 pixel.
            height=700,
            # size_max=65: Kích thước tối đa của điểm là 65.
            size_max=65,
            # log_x=True: Sử dụng tỷ lệ log cho trục x.
            log_x= True,
            # hover_data=data.columns: Hiển thị thông tin khi di chuột qua điểm trên biểu đồ.
            hover_data=data.columns,
            # hover_name='country': Hiển thị tên quốc gia khi di chuột qua điểm.
            hover_name='country',
            # nếu mn k thích giao diện sáng thì mn có thể chuyển sang giao diện màu tối cho biểu đồ bằng cách:
            template='plotly_dark')
figure_scatter.show()
```

Sử dụng tỷ lệ log cho trục x trong biểu đồ phân tán có một số lợi ích:

Biểu diễn dữ liệu rộng rãi: Khi dữ liệu trên trục x có phạm vi rộng, sử dụng tỷ lệ log giúp biểu đồ hiển thị dữ liệu một cách rõ ràng hơn. Các giá trị nhỏ không bị nén quá mức và vẫn có thể quan sát được.

Phân tích dữ liệu phân phối không đều: Trục x thường có phân phối không đều, ví dụ như GDP per capita. Sử dụng tỷ lệ log giúp làm dịu sự chênh lệch giữa các giá trị và giúp phân tích dữ liệu dễ dàng hơn.

Biểu diễn dữ liệu lớn và nhỏ cùng lúc: Tỷ lệ log cho phép biểu diễn dữ liệu lớn và nhỏ cùng lúc trên cùng một trục. Điều này hữu ích khi bạn muốn so sánh các quốc gia có GDP per capita khác biệt lớn.



- mình sẽ thử bỏ parameter log_x= True,thì mn có thể thấy biểu đồ hiển thị rất khó để quan sát


Sự khác biệt giữa biểu đồ Scatter và scatter_geo :

- Loại dữ liệu: Scatter sử dụng tọa độ toán học, trong khi scatter_geo sử dụng tọa độ địa lý.
- Hiển thị: Scatter hiển thị trên hệ tọa độ 2D, còn scatter_geo hiển thị trên bản đồ.

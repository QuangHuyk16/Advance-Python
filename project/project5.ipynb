# **Đồ thị chuỗi thời gian**


Biểu đồ chuỗi thời gian là một biểu đồ đường được sử dụng để trực quan hóa dữ liệu chuỗi thời gian. Dữ liệu chuỗi thời gian là dữ liệu được thu thập trong một khoảng thời gian. Một số ví dụ phổ biến về dữ liệu chuỗi thời gian là dữ liệu giá cổ phiếu, dữ liệu bán hàng hàng ngày, dữ liệu trường hợp mắc bệnh hàng ngày, v.v.

Trong khi hiển thị biểu đồ chuỗi thời gian, các khoảng thời gian nằm trên trục x và các điểm dữ liệu được thu thập trong các khoảng thời gian đó nằm trên trục y.

Ví dụ: trong khi hiển thị biểu đồ chuỗi thời gian của dữ liệu giá cổ phiếu, cột ngày sẽ nằm trên trục x và cột giá sẽ nằm trên trục y

# **Biểu đồ chuỗi thời gian bằng Python**

Để trực quan hóa biểu đồ chuỗi thời gian bằng Python, tôi sẽ sử dụng bộ dữ liệu giá cổ phiếu. Có nhiều thư viện bằng Python để trực quan hóa dữ liệu; Tôi sẽ sử dụng Plotly vì có thể dễ dàng hình dung các hình ảnh trực quan tương tác bằng cách sử dụng Plotly. Vì vậy, hãy bắt đầu nhiệm vụ này bằng cách thu thập dữ liệu giá cổ phiếu mới nhất của Apple:

```python
import pandas as pd
import yfinance as yf
import datetime
from datetime import date, timedelta # nhập các lớp date và timedelta từ thư viện datetime
                                    #Lớp date được sử dụng để đại diện cho một ngày cụ thể trong lịch
                                  #lớp timedelta được sử dụng để tính toán khoảng thời gian giữa 2 ngày
today = date.today()

d1 = today.strftime("%Y-%m-%d") #chuyển đổi đối tượng date today thành một chuỗi đại diện cho ngày hôm nay theo định dạng "YYYY-MM-DD" và gán cho biến d1
end_date = d1

d2 = date.today() - timedelta(days=365)
d2 = d2.strftime("%Y-%m-%d")
start_date = d2

data = yf.download('AAPL',
                      start=start_date,
                      end=end_date,
                      progress=False)#progress=False được sử dụng để tắt thông báo tiến trình tải xuống
data.head()
```

Bây giờ bên dưới là cách bạn có thể hình dung biểu đồ chuỗi thời gian bằng Python:

sử dụng thư viện Plotly Express để vẽ biểu đồ đường (line chart) cho dữ liệu đã tải xuống từ Yahoo! Finance.

```python
import plotly.express as px
figure = px.line(data, x = data.index, y = "Close")
figure.update_layout(title="Dự đoán giá đóng cửa của tài sản XYZ")
figure.show()
```


# Để đọc và hiểu biểu đồ chuỗi thời gian, bạn cần quan sát các thành phần chính sau:

**1**.Trục X (Ngày/Thời gian): Đây là trục ngang của biểu đồ, thường thể hiện thời gian trong đơn vị cụ thể như ngày, tháng, năm hoặc giờ. Trục này biểu diễn thời gian theo thứ tự từ trái qua phải.

**2**.Trục Y (Dữ liệu): Đây là trục dọc của biểu đồ, thường thể hiện các giá trị tương ứng với thời gian trên trục X. Dữ liệu có thể là bất kỳ thứ gì bạn đang theo dõi hoặc đo lường, ví dụ như doanh số bán hàng, nhiệt độ, lượng mưa, v.v.

**3**.Đường/ký hiệu biểu diễn dữ liệu: Đường hoặc các ký hiệu khác trên biểu đồ thể hiện giá trị của dữ liệu tại mỗi điểm thời gian trên trục X. Trong một số trường hợp, có thể sử dụng các loại biểu diễn dữ liệu khác như cột (bar chart), điểm (scatter plot), v.v.

**4**.Tiêu đề: Tiêu đề thường đặt ở phía trên của biểu đồ, giúp cho người đọc hiểu được nội dung của biểu đồ là gì.

**5**.Chú thích (nếu có): Một số biểu đồ có thể đi kèm với chú thích giải thích ý nghĩa của các màu sắc, ký hiệu hoặc dạng dữ liệu trên biểu đồ.

**6**.Dải giá trị trục X và Y: Đôi khi, biểu đồ cũng có thể hiển thị các dải giá trị trên trục X và Y để người đọc có thể xác định giá trị của dữ liệu một cách chính xác.

Khi đọc một biểu đồ chuỗi thời gian, bạn cần chú ý đến các thành phần trên để hiểu được xu hướng, biến động của dữ liệu theo thời gian và các thông tin liên quan.

# Biểu đồ tần suất và biểu đồ mật độ sử dụng Python

### Biểu đồ tần suất và biểu đồ mật độ dùng để hình dung và phân tích sự phân bố của tập dữ liệu. 

Biểu đồ tần suất (Histogram): 
- Thể hiện tần suất xuất hiện của mỗi giá trị khác nhau trong một tập dữ liệu, cho thấy sự thay đổi và biến động của tập dữ liệu.
- Thích hợp cho dữ liệu rời rạc (không liên tục).
- Dễ hiểu và thường được sử dụng trong kiểm soát chất lượng và phân tích dữ liệu.

Điểm mạnh:
- Hiển thị rõ ràng phân bố của dữ liệu.
- Dễ dàng so sánh tần suất giữa các giá trị.
- Cho thấy các điểm ngoại lai (outliers) nếu có.

Điểm yếu:
- Không phù hợp cho dữ liệu liên tục (cần phân chia thành các khoảng).
- Có thể bị ảnh hưởng bởi số lượng khoảng chia.

Biểu đồ mật độ (Density Plot):

- Thể hiện phân phối xác suất của biến ngẫu nhiên, cho thấy mật độ xác suất tại mỗi giá trị.
- Dựa vào hình dạng của biểu đồ, ta có thể đánh giá phân phối của dữ liệu.
- Dạng đường cong liên tục, thích hợp cho dữ liệu liên tục.

Điểm mạnh:
- Hiển thị rõ ràng phân phối của dữ liệu.
- Dễ dàng xem mật độ xác suất tại mỗi giá trị, so sánh hình dạng phân phối giữa các nhóm.
- Không cần phân chia thành khoảng.

Điểm yếu:
- Không thể hiển thị tần suất trực tiếp.

# Cách trực quan hóa biểu đồ tần suất và biểu đồ mật độ bằng Python.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style()
sns.set_context()

data = np.random.multivariate_normal([0, 0], [[5, 2], [2, 2]], size=2000)
data = pd.DataFrame(data, columns=['x', 'y'])
plt.hist(data["x"], alpha=0.5)
plt.hist(data["y"], alpha=0.5)
plt.show()
```

Dưới đây là cách bạn có thể hình dung biểu đồ mật độ bằng Python:

```python
import warnings
warnings.filterwarnings("ignore", category=FutureWarning)

data = np.random.multivariate_normal([0, 0], [[5, 2], [2, 2]], size=2000)
data = pd.DataFrame(data, columns=['x', 'y'])
sns.kdeplot(data["x"], shade=True)
sns.kdeplot(data["y"], shade=True)
plt.show()
```

Chúng ta cũng có thể hình dung cả biểu đồ tần suất và biểu đồ mật độ cùng một lúc. Dưới đây là cách bạn có thể hình dung cả hai bằng Python:

```python
data = np.random.multivariate_normal([0, 0], [[5, 2], [2, 2]], size=2000)
data = pd.DataFrame(data, columns=['x', 'y'])
sns.histplot(data['x'], kde=True, fill=True)
sns.histplot(data['y'], kde=True, fill=True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Tạo dữ liệu ngẫu nhiên
data = np.random.normal(loc=0, scale=1, size=1000)

# Vẽ biểu đồ tần suất (Histogram)
plt.figure(figsize=(8, 6))
plt.hist(data, bins=30, color='skyblue', edgecolor='black', alpha=0.7)
plt.xlabel('Giá trị')
plt.ylabel('Tần suất')
plt.title('Biểu đồ tần suất')
plt.show()

# Vẽ biểu đồ mật độ (Density Plot)
plt.figure(figsize=(8, 6))
sns.kdeplot(data, color='orange', fill=True)
plt.xlabel('Giá trị')
plt.ylabel('Mật độ xác suất')
plt.title('Biểu đồ mật độ')
plt.show()
```
